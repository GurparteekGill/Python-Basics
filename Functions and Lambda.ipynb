# Dynamic arguments (*args)
   Also known as dynamic arguments in args we need not to wwrite the actual number of arguments 
   here we dont need to change the number of parameters everytime we call the function.
   we can do it dynamically


```python
def database_c(*args):#
    for arg in args:
        print(arg)
        
database_c("name","weight","age")
```

```python
def database_c(**kwargs):#
    for value in kwargs.items():
        print(value)
        
database_c(name='Gurparteek',age="45")
```

# Lambda Function
One of Pythons most useful (and for beginners, confusing) tools is the lambda expression. Lambda expressions allow us to create "anonymous" functions. This basically means we can quickly make ad-hoc functions without needing to properly define a function using def.

Function objects returned by running lambda expressions work exactly the same as those created and assigned by defs. There is key difference that makes lambda useful in specialized roles:

Lambda's body is a single expression, not a block of statements.

Python supports the creation of anonymous functions at runtime, using a construct called lambda
This approach is most commonly used when passing a simple function as an argument to another function.
Now Lets slowly break down a lambda expression by deconstructing a function:

syntax:

variable_to_store= lambda arg1,arg2 : operation

```python
def square(num):
    result = num**2
    return result
square(2)
```

```python
square_lambda= lambda  num: num**6
```

```python
square_lambda(2)
```

```python
subtraction_lambda= lambda number1,number2 : number1 - number2 
```

```python
subtraction_lambda(45,23)
```
